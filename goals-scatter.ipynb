```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

tabell = pd.read_html('https://www.altomfotball.no/element.do?cmd=tournament&tournamentId=230&useFullUrl=false')
```

```python
print(tabell[2])
```

```python
data = tabell[2]
```

```python
data.head()
```

```python
plt.rcParams['figure.figsize'] = [12, 12] # endrer default figsize i jupyter

x = data["+"] / data["K"] # mål skåret / antall kamper spilt
y = data["-"] / data["K"] # mål innsluppet / antall kamper spilt

n = data["Lag"] # liste med lag


plt.title("Mål skåret vs innsluppet", fontsize=24)
plt.xlabel("Snitt skåret", fontsize=16)
plt.ylabel("Snitt innsluppet", fontsize=16)

plt.plot(x, y, 'ro')

for lag in range(len(n)):
    plt.text(x[lag] + 0.01, y[lag] + 0.01, n[lag], fontsize=14)

plt.grid(linestyle="--", linewidth=0.9)
plt.show()
```
